```python
import numpy as np
import matplotlib.pyplot as plt
```

## Example01

```python
T = np.loadtxt("example01/output/conductance_bulk.dat")
T_ref = np.loadtxt("../tests/reference/Fortran/example01/conductance_bulk.dat")
plt.plot(T[:, 0], T[:, 1], label="Python")
plt.plot(T_ref[:, 0], T_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 1: Conductance (Bulk)")
plt.xlabel("Energy (eV)")
plt.legend()
plt.show()

dos = np.loadtxt("example01/output/doscond_bulk.dat")
dos_ref = np.loadtxt("../tests/reference/Fortran/example01/doscond_bulk.dat")
plt.plot(dos[:, 0], dos[:, 1], label="Python")
plt.plot(dos_ref[:, 0], dos_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 1: DOS")
plt.legend()
plt.show()
```

## Example02

```python
T = np.loadtxt("example02/output/conductance_bulk.dat")
T_ref = np.loadtxt("../tests/reference/Fortran/example02/conductance_bulk.dat")
plt.plot(T[:, 0], T[:, 1], label="Python")
plt.plot(T_ref[:, 0], T_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 2: Conductance (Bulk)")
plt.legend()
plt.show()

for i in range(2, T.shape[1]):
    plt.plot(T[:, 0], T[:, i], label="Python")
    plt.plot(T_ref[:, 0], T_ref[:, i], '-.' , label="Fortran ref")
    plt.title(f"Example 2: Conductance (Bulk) Channel {i-1}")
    plt.legend()
    plt.show()
T = np.loadtxt("example02/output/conductance_lcr.dat")
T_ref = np.loadtxt("../tests/reference/Fortran/example02/conductance_lcr.dat")
plt.plot(T[:, 0], T[:, 1], label="Python")
plt.plot(T_ref[:, 0], T_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 2: Conductance (LCR)")
plt.legend()
plt.show()
T = np.loadtxt("example02/output/conductance_lead.dat")
T_ref = np.loadtxt("../tests/reference/Fortran/example02/conductance_lead.dat")
plt.plot(T[:, 0], T[:, 1], label="Python")
plt.plot(T_ref[:, 0], T_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 2: Conductance (Lead)")
plt.legend()
plt.show()

dos = np.loadtxt("example02/output/doscond_bulk.dat")
dos_ref = np.loadtxt("../tests/reference/Fortran/example02/doscond_bulk.dat")
plt.plot(dos[:, 0], dos[:, 1], label="Python")
plt.plot(dos_ref[:, 0], dos_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 2: DOS")
plt.legend()
plt.show()
dos = np.loadtxt("example02/output/doscond_lcr.dat")
dos_ref = np.loadtxt("../tests/reference/Fortran/example02/doscond_lcr.dat")
plt.plot(dos[:, 0], dos[:, 1], label="Python")
plt.plot(dos_ref[:, 0], dos_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 2: DOS (LCR)")
plt.legend()
plt.show()
dos = np.loadtxt("example02/output/doscond_lead.dat")
dos_ref = np.loadtxt("../tests/reference/Fortran/example02/doscond_lead.dat")
plt.plot(dos[:, 0], dos[:, 1], label="Python")
plt.plot(dos_ref[:, 0], dos_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 2: DOS (Lead)")
plt.legend()
plt.show()

current = np.loadtxt("example02/output/current.dat")
current_ref = np.loadtxt("../tests/reference/Fortran/example02/current.dat")
plt.plot(current[:, 0], current[:, 1], label="Python")
plt.plot(current_ref[:, 0], current_ref[:, 1], '-.' , label="Fortran ref")
plt.title("Example 2: Current")
plt.legend()
plt.show()
```
